```python
# Clean industry column
top_1000_companies[top_1000_companies['industry'].isnull()]

top_1000_companies.loc[210, 'industry'] = 'Hospitality'

top_1000_companies.loc[280, 'industry'] = 'Manufacturing'

top_1000_companies.loc[531, 'industry'] = 'Restaurants'

top_1000_companies['industry'] = top_1000_companies['industry'].str.strip()

#Change the datatype of the funding column


industry_categories = {
    "Technology": [
        "Cloud", "DevOps", "Wireless", "Electronics", "Martech", "3D", "IT Security", "AI",
        "Digital Health", "Robotics", "AdTech", "Adtech", "Blockchain", "Cybersecurity", "Internet",
        "SaaS", "Babytech", "Program Dev" , "CRM", "AI", "Printing",
        "Analytics", "Data Analytics", "Salestech" , "Tech Services", "Design", "HRTech",
        "Telecom", "Environmental", "BabyTech", "Real Estate Tech",
         "Software", "Hardware", "NoCode/LowCode", "Information Technology & Services"
    ],
    "Finance & Fintech": [
        "Fintech", "Finance", "Banking", "Investments", "InsureTech", "Cryptocurrency", "Insurance", "Fund-Raising", "FinTech"
    ],
    "E-commerce & Retail": [
        "eCommerce", "Retail", "eCommerceTech", "Furniture and Home Furnishings Manufacturing", "Apparel", "Supplies",
        "Restaurants", "Consumer", "Cosmetics", "Trade", "Food"
    ],
    "Healthcare & Pharmaceuticals": [
        "Digital Health", "Hospital/Healthcare", "Pharma", "Medical Equip", "Pharmaceuticals", "Hospital/Health", "Cannabis",
        "Biotech", "Health", "Medical Offices", "Veterinary"
    ],
    "Manufacturing & Industrial": [
        "Energy/Oil", "Industrial", "Automation Machinery Manufacturing", "Machinery", "Packaging and Containers Manufacturing",
        "Materials", "Mining", "Manufacturing", "Chemicals", "Construction", "Automotive", "Textiles", "Warehousing", "Farming", "Semiconductors",
        "Automation Machinery Manufacturing", "Energy",  "Oil and Gas", "Engineering"
    ],
    "Transportation & Travel": [
        "Aviation", "Airlines and Aviation", "Logistics", "Transportation", "Travel Arrangements", "Delivery", "DeliveryTech"
    ],
    "Media & Entertainment": [
        "Media", "Entertainment", "Video", "Casinos", "Gaming", "Sports", "Newspapers Publishing", "Marketing", "Leisure",
        "Newspaper Publishing", "Advertising", "Hospitality"
    ],
    "Education & Training": [
        "EdTech", "Education", "E-Learning Providers", "Business Skills Training", "Edtech", "Education"
    ],
    "Consulting & Professional Services": [
        "Legaltech", "LegalTech", "Consulting", "Outsourcing and Offshoring Consulting", "Consulting & Professional Services", 
        "Think Tanks", "Research", "HR", "Executive", "Support/CRM Tech", "Nonprofit", "Recruiting", "Real Estate", "Security",
         "Business Intelligence", "Networking", "Defense"]
}



def map_general_industry(industry):
    for category, industries in industry_categories.items():
        if industry in industries:
            return category
    return 'Other'


top_1000_companies['general_industry'] = top_1000_companies['industry'].apply(map_general_industry)
```

```python
top_1000_companies[top_1000_companies['general_industry'] == 'Other']['industry'].unique()
```

```python
# clean growth percentage

top_1000_companies['growth_percentage'] = top_1000_companies['growth_percentage'].apply(lambda x: float(x[:-1]))
```

```python
# clean funding column
#Fix this function, if it starts with the euro sign multiply by the conversion of 1000000 into euros
#If CA, multiply by the canadian dollar conversion of 100,000,000 into euros 

#First check if the euro sign is in the front. Check if CA is in the string

def clean_funding(fund):
    if isinstance(fund, str):
        clean_fund = fund[1:]
        if fund[0] == 'â‚¬':
            clean_fund_euro = fund[1:]
            if 'M' in  clean_fund_euro:
                fund_euro_million = float(clean_fund_euro[:-1]) * 1085105
                return fund_euro_million
        if fund[0:2] == 'CA':
            clean_fund_canada = fund[3:]
            if 'M' in clean_fund_canada:
                fund_canada_million = float(clean_fund_canada[:-1]) * 737630
                return fund_canada_million                 
        elif 'M' in clean_fund:
            fund_million = float(clean_fund[:-1]) * 1000000
            return fund_million
        elif 'B' in clean_fund:
            fund_billion = float(clean_fund[:-1]) * 1000000000
            return fund_billion
    else:
        return fund

top_1000_companies['funding'] = top_1000_companies['funding'].apply(clean_funding)
        

top_1000_companies.head()
```

```python
states_grouped = pd.read_sql(''' SELECT state,
                             COUNT(state) AS state_count,
                             100.0 * COUNT(state)/ (SELECT COUNT(*) FROM top_1000_companies) AS pct_companies
                             FROM top_1000_companies
                             WHERE state  IS NOT 'Unlisted'
                             GROUP BY state
                             ORDER BY state_count DESC''',
                             companies_conn
                             )
states_grouped
```
